## [iSIP](https://www.ifap.pt/isip/ows/)

- For 2023 and 2024 use Culturas_PTC.gpkg (Portugal Continental, culturas.zip) and 2023_2024_tables.gpkg (Estrutura do Servico, info_help.zip)

```python
# get familiar with the structure:

#get modules
import geopandas as gdp
import fiona

#paths 
field_level_file = "../Culturas_PTC.gpkg"
lu_file = "../2023_2024_tables.gpkg"

#field_shapes
layers = fiona.listlayers(field_level_file)

#lu field layer
layers_lu = fiona.listlayers(lu_file)

# #load specific layer:
gdf = gdp.read_file(field_level_file, layer="Culturas_PC_2023")
gdf_lu = gdp.read_file(lu_file, layer="c_castelo_branco_2023")

#are the ids unique?
print(len(gdf_lu['CUL_ID'].unique())) # field id
print(len(gdf_lu['OSA_ID'].unique())) # sub parcel id --> do not fully understand this one, why less?

#create list of only Almond ids
print(gdf_lu.columns) #column names
gdf_lu_almonds = gdf_lu[gdf_lu['PUN_CUL_COD']==109]
almond_fields_lst = gdf_lu_almonds['CUL_ID'].to_list()

# #select field geometries:
almond_fields = gdf[gdf['CUL_ID'].isin(almond_fields_lst)]
almond_fields.to_file("../almonds_2023.geojson", driver="GeoJSON")
```
